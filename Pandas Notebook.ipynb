### Using Agg method
##### There's a need to inport our np to complete this
##### The agg helps us to bring every of our math together in thesame column and row well defined and istead of performing them one after the other

```python
walmart[['Weekly_Sales', 'Temperature', 'Fuel_Price']].agg(np.median)
```

### Interquantile  WITH Agg

```python
# create our function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
```

```python
# Displaying the result of median and interquantile

median_interquantile = walmart[['Weekly_Sales', 'Temperature', 'Fuel_Price']].agg([iqr,np.median])
median_interquantile
```
